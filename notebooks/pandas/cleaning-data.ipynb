```python
df = pd.DataFrame(np.arange(12).reshape((3, 4)), columns=list('abcd'))
```

```python
df.loc[1, 'd'] = np.nan
df.loc[2, 'c'] = np.nan
```

```python
df.dropna(axis=1)
```

```python
df
```

```python
df.fillna(0)
```
